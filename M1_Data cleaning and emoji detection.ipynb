```python
reviews1 = reviews_trans[['App Name','Review ID', 'Rating','Country', "Subject", 'Body', "Translated Subject", 'Translated Body',\
                         'lang', 'conf', 'trans']]
```

```python
df_diana = pd.read_csv('trans_China_1.csv')
df_diana.set_index('Index', inplace=True)
df_diana.head()
```

```python
df_diana.drop('Unnamed: 0',axis=1,inplace = True)
df_trans = reviews1.merge(df_diana, how='left',left_index=True, right_index=True)
df_trans.head()
```

```python
#write function that won't concat if one of two columns is null
import numpy as np
def concat(x, y):
    if x != x:  # this means x is NaN
        return y
    if y != y:
        return x
    return x + ' ' + y
```

```python
df_trans['trans_Subj_Body'] = df_trans.apply(lambda x: concat(x['Translated Subject'], x['Translated Body']),axis=1)
```

```python
df_trans['trans'] = df_trans.apply(lambda x: concat(x['trans_x'], x['trans_y']), axis=1)
```

```python
df_trans['trans_Subj_Body'].isnull().sum()
```

```python
df_trans["all_text"] = df_trans.apply(lambda x: concat(x['trans_Subj_Body'], x['trans']),axis=1)
```

```python
df_trans["all_text"].isnull().sum()
```

```python
# APPLY eMOJI EXTRACTION

import demoji
demoji.download_codes()
```

```python
def find_emoji(review):
    '''This function will extract emoji from reviews'''
    return demoji.findall(str(review))
def remove_emoji(review):
    return demoji.replace(str(review))
```

```python
df_trans["all_text_emoji"] = df_trans["all_text"].apply(lambda x: find_emoji(x))
```

```python
df_trans["all_text_NOemoji"] = df_trans["all_text"].apply(lambda x: remove_emoji(x))
```

```python
df_trans.head()
```

```python
df_trans.loc[df_trans["all_text_emoji"] != {}, 'all_text_emoji']
```

```python
#clean function
import re
my_punctuation = '!"$%&\'()*+,-./:;<=>?[\\]^_`{|}~•@#'

def clean_text(text):
    text = re.sub('['+my_punctuation + ']+', ' ', str(text)) # strip punctuation
    text = re.sub('nan+', ' ', str(text)) # remove nan
    text = re.sub('NaN+', ' ', str(text)) # remove nan
    text = re.sub('([0-9]+)', '', str(text)) # remove numbers
    text = re.sub('\s+', ' ', str(text)) #remove double spacing
    text = text.lower() # lower case
    return text
```

```python
df_trans["cleaned_text"]=df_trans['all_text_NOemoji'].apply(lambda x: clean_text(x))
```

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from collections import Counter
```

```python
combined_csv_all = pd.read_csv("Combined_Data.csv")
```

```python
combined_csv_all['App Name'].value_counts()
```

```python
combined_data = combined_csv_all.set_index('App Name').loc[['Maps - Navigation & Transit','Waze - GPS, Maps, Traffic Alerts & Live Navigation',\
                                 'Google Maps - GPS Navigation','Waze Navigation & Live Traffic',\
                                 'HERE WeGo - Offline Maps & GPS','HERE WeGo - City navigation'], :]
```

```python
combined_data.reset_index(inplace = True)
```

```python
combined_data['App Store'].value_counts()
```

```python
combined_data['App Name'].value_counts()
```

```python
combined_data['App Name'].replace({'Maps - Navigation & Transit':'Google Maps' , 'Google Maps - GPS Navigation':'Google Maps',
                'Waze - GPS, Maps, Traffic Alerts & Live Navigation':'Waze', 
                'Waze Navigation & Live Traffic':'Waze',
                'HERE WeGo - Offline Maps & GPS': 'HERE WeGo','HERE WeGo - City navigation': 'HERE WeGo'}, inplace = True)
```

```python
combined_data['App Name'].value_counts()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Draw Plot
fig, axes = plt.subplots(1, 2, figsize=(20,7), dpi= 80)
combined_data['Country'].value_counts().head(20).plot(kind = 'bar',ax=axes[0])
combined_data['App Name'].value_counts().plot(kind = 'bar')

# Set Title
axes[0].set_title('Top 20 Countries', fontsize=18); 
axes[1].set_title('Number of reviews per each APP', fontsize=18)
plt.show()
```

```python
# Draw Plot
fig, axes = plt.subplots(1, 2, figsize=(20,7), dpi= 80)
sns.boxplot(x='App Name', y='Rating', data=combined_data, ax=axes[0])
combined_data['Rating'].value_counts().plot(kind = 'bar')

# Set Title
axes[0].set_title('Rating range across different APP', fontsize=18); 
axes[1].set_title('Ratings values across all reviews', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
fig = combined_data['Country'].value_counts().head(20).plot(kind = 'bar')
fig.set_title('Top 20 Countries')
fig.set_xlabel('Country')
fig.set_ylabel('Frequency')
#fig.set_xticklabels(x_labels)
```

```python
plt.figure(figsize=(12, 8))
fig = combined_data['App Name'].value_counts().plot(kind = 'bar')
fig.set_title('Number of reviews per each APP')
fig.set_xlabel('Application Name')
fig.set_ylabel('Frequency')
#fig.set_xticklabels(x_labels)
```

```python
plt.figure(figsize=(12, 8))
fig = combined_data['Rating'].value_counts().plot(kind = 'bar')
fig.set_title('Ratings values across all reviews')
fig.set_xlabel('Rating value')
fig.set_ylabel('Frequency')
#fig.set_xticklabels(x_labels)
```

```python
plt.figure(figsize=(12, 8))

fig = sns.boxplot(x='App Name', y='Rating', data=combined_data)
fig.set_title('Rating range across different APP')
fig.set_xlabel('Application name')
fig.set_ylabel('Rating')
```
